**1. Penentuan data yang digunakan dalam Visualisasi:**

      Kita akan menggunakan dataset "diabetes_preprocessed.csv" yang berisi beberapa atribut berikut:
      -Glucose: Tingkat glukosa pada pasien
      -BloodPressure: Tekanan darah
      -SkinThickness: Ketebalan kulit
      -Insulin: Kadar insulin dalam darah
      -BMI: Body Mass Index (Indeks Massa Tubuh)
      -DiabetesPedigreeFunction: Fungsi yang mengukur risiko diabetes berdasarkan riwayat keluarga
      -Age: Usia pasien
      -Outcome: Hasil diagnosis apakah pasien menderita diabetes (1) atau tidak (0)

**2. Tujuan, User, Fungsi, dan Tone dari Visualisasi:**

    -Tujuan: Menunjukkan distribusi atribut kesehatan pasien diabetes dan mengamati hubungan antara berbagai variabel, seperti glukosa,
    tekanan darah, insulin, dan BMI, untuk membantu dalam memahami faktor-faktor yang berhubungan dengan diabetes.
    -User: Peneliti atau dokter yang ingin menganalisis risiko dan kondisi diabetes pada pasien.
    -Fungsi: Memvisualisasikan pola distribusi dan hubungan antara atribut kesehatan untuk membantu identifikasi faktor risiko diabetes.
    -Tone: Sederhana dan informatif, dengan warna yang berbeda untuk membedakan variabel usia pasien dan memberikan wawasan yang mudah dipahami.

**3. Visualisasi Data**

```python
# Import library yang dibutuhkan
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Mengimpor dataset diabetes
df = pd.read_csv('diabetes_preprocessed.csv')
```

```python
# 1. Visualisasi distribusi tingkat glukosa
plt.figure(figsize=(8, 6))
sns.histplot(df['Glucose'], kde=True, color='skyblue')
plt.title('Distribusi Tingkat Glukosa pada Pasien Diabetes')
plt.xlabel('Tingkat Glukosa')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# 2. Visualisasi distribusi tekanan darah
plt.figure(figsize=(8, 6))
sns.histplot(df['BloodPressure'], kde=True, color='green')
plt.title('Distribusi Tekanan Darah pada Pasien Diabetes')
plt.xlabel('Tekanan Darah')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# 3. Visualisasi distribusi BMI
plt.figure(figsize=(8, 6))
sns.histplot(df['BMI'], kde=True, color='purple')
plt.title('Distribusi BMI pada Pasien Diabetes')
plt.xlabel('BMI')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# 4. Visualisasi hubungan antara Glukosa, Insulin, dan BMI
plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(x='Glucose', y='BMI', hue='Insulin', data=df, palette='coolwarm')
plt.title('Hubungan Glukosa, BMI, dan Insulin')
plt.xlabel('Tingkat Glukosa')
plt.ylabel('BMI')
plt.legend(title='Insulin')
plt.show()
```

```python
# 5. Visualisasi hubungan antara Glukosa, DiabetesPedigreeFunction, dan Usia
plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(x='Glucose', y='DiabetesPedigreeFunction', hue='Age', data=df, palette='coolwarm')
plt.title('Hubungan Glukosa, Fungsi Pedigree Diabetes, dan Usia')
plt.xlabel('Tingkat Glukosa')
plt.ylabel('Diabetes Pedigree Function')
plt.legend(title='Usia')
plt.show()
```

```python
# 6. Visualisasi hubungan Outcome (diabetes atau tidak) dengan BMI dan Glukosa
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=df, palette='Set1')
plt.title('Hubungan Glukosa, BMI, dan Outcome (Apakah Menderita Diabetes)')
plt.xlabel('Tingkat Glukosa')
plt.ylabel('BMI')
plt.legend(title='Outcome (1=Diabetes, 0=Tidak)')
plt.show()
```
